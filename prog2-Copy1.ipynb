```python
import matplotlib.pyplot as plt
import numpy as np

X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)

plt.plot(X, C, color="blue", linewidth=1, label='cos')
plt.plot(X, S, color="red",  linewidth=2.5, linestyle="--", label='sin')

plt.xticks([-2*np.pi, -np.pi/2, 0, np.pi/2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$23$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, +1])

plt.legend()
plt.show() 
```
